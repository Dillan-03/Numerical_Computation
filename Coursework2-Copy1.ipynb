```python
#Question 5.3 Convergence Testing

import math
import numpy as np
import matplotlib.pyplot as plt

def Euler_method(T, theta, v_o, dt):
    n = int(6 * (math.pi) / dt)
    
    t = np.zeros([n + 1, 1])        # Initializing array time
    Theta = np.zeros([n + 1, 1])    # Initializing the array of angles Theta
    v = np.zeros([n + 1, 1])        # Initializing the array velocity

    # Initializing the first element of each respective variables
    t[0] = 0.0                           
    Theta[0] = theta
    v[0] = v_o

    # Calculating all the angles, velocities, and time for each i
    for i in range(n):
        t[i + 1] = t[i] + dt
        Theta[i + 1] = Theta[i] + dt * (v[i])
        v[i + 1] = v[i] + dt * (-math.sin(Theta[i]))
        
    
    plt.title("Theta/t")
    plt.xlabel("x", loc='right')
    plt.ylabel ("y", loc='top')
    plt.grid()
    plt.plot(t, Theta)
   

Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 20)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 40)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 80)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 160)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 320)
plt.legend(['6pi/20', '6pi/40', '6pi/80', '6pi/160', '6pi/320'])
plt.show()

Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 200)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 400)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 800)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 1600)
Euler_method(6 * (math.pi), 0, 1, 6 * (math.pi) / 3200)
plt.legend(['6pi/200', '6pi/400', '6pi/800', '6pi/1600', '6pi/3200'])
plt.show()
```
